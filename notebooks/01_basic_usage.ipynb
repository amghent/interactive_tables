# Interactive Table examples - Part 1: Basic usage

## Purpose

This notebook and the other notebooks in this directory contain examples on how to use the configuration module.

## Setup of the dataframe

```python
import pandas as pd
import os
from pathlib import Path

notebook_dir = Path(os.getcwd())
data_dir = os.path.join(notebook_dir.parent, "world_cities")
data_file = os.path.join(data_dir, "worldcities.csv")

cities = pd.read_csv(data_file)
cities = cities[cities["iso2"]=="BE"]
```

## show() method

```python
from interactive_tables.interactive_table import InteractiveTable

cities_table = InteractiveTable(dataframe=cities)
cities_table.show()
```

## Shortcut method

```python
from interactive_tables.interactive_table import InteractiveTable

InteractiveTable(cities).show()
```

## head() method


```python
from interactive_tables.interactive_table import InteractiveTable

cities_table = InteractiveTable(dataframe=cities)

cities_table.head(15)
cities_table.show()
```

## drop_index() method


```python
from interactive_tables.interactive_table import InteractiveTable

cities_table = InteractiveTable(dataframe=cities)

cities_table.drop_index()
cities_table.show()
```

## create_index() method

```python
from interactive_tables.interactive_table import InteractiveTable

cities_table = InteractiveTable(dataframe=cities)

cities_table.create_index(start_value=1)
cities_table.show()
```

```python
from interactive_tables.interactive_table import InteractiveTable

cities_table = InteractiveTable(dataframe=cities)

cities_table.drop_index()
cities_table.create_index()
cities_table.show()
```
